```python
def count_ways_to_construct_designs(input_data):
    # Split the input data by lines
    lines = input_data.strip().split('\n')
    
    # The first line contains the towel patterns, split them by commas
    towel_patterns = [pattern.strip() for pattern in lines[0].split(',')]
    
    # The remaining lines are the designs
    designs = [design.strip() for design in lines[1:]]
    
    # Convert towel patterns to a set for quick lookup
    pattern_set = set(towel_patterns)
    
    def count_ways(design):
        n = len(design)
        dp = [0] * (n + 1)
        dp[0] = 1  # There's one way to construct an empty string
        
        for i in range(1, n + 1):
            for j in range(i):
                if design[j:i] in pattern_set:
                    dp[i] += dp[j]
        
        return dp[n]
    
    total_ways = 0
    
    for design in designs:
        ways = count_ways(design)
        total_ways += ways
        print(f"Design '{design}' can be constructed in {ways} ways.")
    
    return total_ways

# Example input as a multi-line string (replace with your actual puzzle input)
input_data = """
r, wr, b, g, bwu, rb, gb, br
brwrr
bggr
gbbr
rrbgbr
ubwu
bwurrg
brgr
bbrgwb
"""

with open("Input/InputDay19P1.txt", "r") as f:
    input_data = f.read()

# Calculate the total number of ways to construct all designs
result = count_ways_to_construct_designs(input_data)
print(f"Total number of ways to construct all designs: {result}")
```
