```python
def count_possible_designs(towel_patterns, designs):
    # Convert towel patterns to a set for quick lookup
    pattern_set = set(towel_patterns)
    
    def can_construct_design(design):
        n = len(design)
        dp = [False] * (n + 1)
        dp[0] = True
        
        for i in range(1, n + 1):
            for j in range(i):
                if dp[j] and design[j:i] in pattern_set:
                    dp[i] = True
                    break
        
        return dp[n]
    
    possible_count = 0
    
    for design in designs:
        if can_construct_design(design):
            possible_count += 1
    
    return possible_count

# Example input
towel_patterns = ["r", "wr", "b", "g", "bwu", "rb", "gb", "br"]
designs = [
    "brwrr",
    "bggr",
    "gbbr",
    "rrbgbr",
    "ubwu",
    "bwurrg",
    "brgr",
    "bbrgwb"
]

# Calculate the number of possible designs
result = count_possible_designs(towel_patterns, designs)
print(result)  # Output: 6
```
