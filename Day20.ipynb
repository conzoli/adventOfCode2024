```python
from collections import deque

# Define directions for movement (up, down, left, right)
DIRECTIONS = [(0, 1), (1, 0), (0, -1), (-1, 0)]

def parse_input(racetrack):
    """Parse the racetrack input into a grid and find start and end positions."""
    grid = [list(line) for line in racetrack.strip().split("\n")]
    start, end = None, None
    for r, row in enumerate(grid):
        for c, cell in enumerate(row):
            if cell == 'S':
                start = (r, c)
            elif cell == 'E':
                end = (r, c)
    return grid, start, end

def is_within_bounds(grid, r, c):
    """Check if a position is within bounds."""
    return 0 <= r < len(grid) and 0 <= c < len(grid[0])

def bfs(grid, start, end):
    """Perform BFS to find the shortest path without cheating."""
    queue = deque([(start, 0)])
    visited = set([start])

    while queue:
        (r, c), steps = queue.popleft()
        if (r, c) == end:
            return steps

        for dr, dc in DIRECTIONS:
            nr, nc = r + dr, c + dc
            if is_within_bounds(grid, nr, nc) and grid[nr][nc] != '#' and (nr, nc) not in visited:
                visited.add((nr, nc))
                queue.append(((nr, nc), steps + 1))
    return float('inf')  # No path found

def bfs_with_cheat(grid, start, end):
    """Perform BFS considering a single 2-picosecond cheat."""
    queue = deque([(start, 0, False)])  # (position, steps, used_cheat)
    visited = set([(start, False)])

    while queue:
        (r, c), steps, used_cheat = queue.popleft()
        if (r, c) == end:
            return steps

        for dr, dc in DIRECTIONS:
            nr, nc = r + dr, c + dc
            if is_within_bounds(grid, nr, nc):
                if grid[nr][nc] != '#' and (nr, nc, used_cheat) not in visited:
                    visited.add((nr, nc, used_cheat))
                    queue.append(((nr, nc), steps + 1, used_cheat))
                elif grid[nr][nc] == '#' and not used_cheat:  # Use cheat
                    for skip_dr, skip_dc in DIRECTIONS:  # Skip walls
                        sr, sc = nr + skip_dr, nc + skip_dc
                        if is_within_bounds(grid, sr, sc) and grid[sr][sc] != '#' and (sr, sc, True) not in visited:
                            visited.add((sr, sc, True))
                            queue.append(((sr, sc), steps + 2, True))

    return float('inf')

def count_large_cheats(racetrack, threshold):
    """Count the number of cheats that save at least 'threshold' picoseconds."""
    grid, start, end = parse_input(racetrack)
    base_time = bfs(grid, start, end)

    if base_time == float('inf'):
        return 0  # No valid path even without cheating

    # Explore all possible cheats
    cheats = []
    for r, row in enumerate(grid):
        for c, cell in enumerate(row):
            if cell == '#':
                # Temporarily remove the wall and calculate new path
                grid[r][c] = '.'
                cheat_time = bfs(grid, start, end)
                grid[r][c] = '#'  # Restore the wall

                if cheat_time < base_time:
                    cheats.append(base_time - cheat_time)

    # Count cheats that save at least the threshold
    return sum(1 for save in cheats if save >= threshold)

# Example racetrack input
racetrack = """###############
#...#...#.....#
#.#.#.#.#.###.#
#S#...#.#.#...#
#######.#.#.###
#######.#.#...#
#######.#.###.#
###..E#...#...#
###.#######.###
#...###...#...#
#.#####.#.###.#
#.#...#.#.#...#
#.#.#.#.#.#.###
#...#...#...###
###############"""

# Threshold for the cheats
threshold = 100

# Count the number of cheats saving at least the threshold
result = count_large_cheats(real_input, threshold)
print(f"Number of cheats saving at least {threshold} picoseconds: {result}")
```

```python
from collections import deque

# Define directions for movement (up, down, left, right)
DIRECTIONS = [(0, 1), (1, 0), (0, -1), (-1, 0)]

def parse_input(racetrack):
    """Parse the racetrack input into a grid and find start and end positions."""
    grid = [list(line) for line in racetrack.strip().split("\n")]
    start, end = None, None
    for r, row in enumerate(grid):
        for c, cell in enumerate(row):
            if cell == 'S':
                start = (r, c)
            elif cell == 'E':
                end = (r, c)
    return grid, start, end

def is_within_bounds(grid, r, c):
    """Check if a position is within bounds."""
    return 0 <= r < len(grid) and 0 <= c < len(grid[0])

def bfs(grid, start, end):
    """Perform BFS to find the shortest path without cheating."""
    queue = deque([(start, 0)])
    visited = set([start])

    while queue:
        (r, c), steps = queue.popleft()
        if (r, c) == end:
            return steps

        for dr, dc in DIRECTIONS:
            nr, nc = r + dr, c + dc
            if is_within_bounds(grid, nr, nc) and grid[nr][nc] != '#' and (nr, nc) not in visited:
                visited.add((nr, nc))
                queue.append(((nr, nc), steps + 1))
    return float('inf')  # No path found

def bfs_with_cheat(grid, start, end, max_cheat_duration):
    """Perform BFS considering a cheat duration up to max_cheat_duration."""
    queue = deque([(start, 0, 0)])  # (position, steps, cheat_used)
    visited = set([(start, 0)])
    min_steps = float('inf')

    while queue:
        (r, c), steps, cheat_used = queue.popleft()
        if (r, c) == end:
            min_steps = min(min_steps, steps)
            continue

        for dr, dc in DIRECTIONS:
            nr, nc = r + dr, c + dc
            if is_within_bounds(grid, nr, nc):
                if grid[nr][nc] != '#' and (nr, nc, cheat_used) not in visited:
                    visited.add((nr, nc, cheat_used))
                    queue.append(((nr, nc), steps + 1, cheat_used))
                elif grid[nr][nc] == '#' and cheat_used < max_cheat_duration:
                    for skip in range(1, max_cheat_duration - cheat_used + 1):
                        sr, sc = r + dr * skip, c + dc * skip
                        if is_within_bounds(grid, sr, sc):
                            if grid[sr][sc] != '#' and (sr, sc, cheat_used + skip) not in visited:
                                visited.add((sr, sc, cheat_used + skip))
                                queue.append(((sr, sc), steps + skip, cheat_used + skip))
    return min_steps

def count_large_cheats_with_extended_rule(racetrack, threshold, max_cheat_duration):
    """Count the number of cheats that save at least 'threshold' picoseconds with extended cheat rules."""
    grid, start, end = parse_input(racetrack)
    base_time = bfs(grid, start, end)

    if base_time == float('inf'):
        return 0  # No valid path even without cheating

    # Calculate path time with extended cheat rule
    cheat_time = bfs_with_cheat(grid, start, end, max_cheat_duration)

    # Determine if the cheat saves enough time
    saved_time = base_time - cheat_time
    return 1 if saved_time >= threshold else 0

# Example racetrack input
racetrack = """###############
#...#...#.....#
#.#.#.#.#.###.#
#S#...#.#.#...#
#######.#.#.###
#######.#.#...#
#######.#.###.#
###..E#...#...#
###.#######.###
#...###...#...#
#.#####.#.###.#
#.#...#.#.#...#
#.#.#.#.#.#.###
#...#...#...###
###############"""

# Threshold for the cheats
threshold = 100
max_cheat_duration = 20

# Count the number of cheats saving at least the threshold
result = count_large_cheats_with_extended_rule(real_input, threshold, max_cheat_duration)
print(f"Number of cheats saving at least {threshold} picoseconds: {result}")
```

```python
from collections import deque
from typing import Generator

Grid = list[list[str]]
Position = tuple[int, int]
DIRECTIONS = ((-1, 0), (1, 0), (0, -1), (0, 1))


def find_point(grid: Grid, type: str) -> Position:
    return next(find_points(grid, type))


def find_points(grid: Grid, type: str) -> Generator[Position, None, None]:
    for row in range(len(grid)):
        for col in range(len(grid[0])):
            if grid[row][col] == type:
                yield row, col


def calculate_distances(
    grid: Grid,
    start_row: int,
    start_col: int,
    walls: set[Position],
) -> dict[Position, int]:
    distances: dict[Position, int] = {}
    queue = deque([(start_row, start_col, 0)])
    seen = set() | walls

    while queue:
        r, c, dist = queue.popleft()

        distances[(r, c)] = dist

        for dr, dc in DIRECTIONS:
            new_r, new_c = r + dr, c + dc

            if (
                0 <= new_r < len(grid)
                and 0 <= new_c < len(grid[0])
                and (new_r, new_c) not in seen
            ):
                seen.add((r, c))
                queue.append((new_r, new_c, dist + 1))

    return distances


def find_possible_shortcuts(
    grid: Grid,
    distances_from_start: dict[Position, int],
    distances_to_end: dict[Position, int],
    walls: set[Position],
    regular_time: int,
    max_cheat_time: int,
) -> list[int]:
    savings = []
    rows, cols = len(grid), len(grid[0])

    for start_r in range(rows):
        for start_c in range(cols):
            if (start_r, start_c) in walls:
                continue

            for end_r in range(
                max(0, start_r - max_cheat_time),
                min(rows, start_r + max_cheat_time + 1),
            ):
                for end_c in range(
                    max(0, start_c - max_cheat_time),
                    min(cols, start_c + max_cheat_time + 1),
                ):
                    if (end_r, end_c) in walls:
                        continue

                    manhattan_dist = abs(end_r - start_r) + abs(end_c - start_c)
                    if manhattan_dist > max_cheat_time:
                        continue

                    total_time = (
                        distances_from_start[(start_r, start_c)]
                        + manhattan_dist
                        + distances_to_end[(end_r, end_c)]
                    )

                    if total_time < regular_time:
                        savings.append(regular_time - total_time)

    return savings


def find_cheats(
    grid: Grid,
    start_row: int,
    start_col: int,
    end_row: int,
    end_col: int,
    walls: set[Position],
    max_cheat_time: int = 2,
) -> list[int]:
    distances_from_start = calculate_distances(grid, start_row, start_col, walls)
    distances_to_end = calculate_distances(grid, end_row, end_col, walls)

    regular_time = distances_from_start[(end_row, end_col)]

    return find_possible_shortcuts(
        grid,
        distances_from_start,
        distances_to_end,
        walls,
        regular_time,
        max_cheat_time,
    )


def part_1(file: str = "Input/InputDay20P1.txt") -> int:
    grid = [list(line) for line in open(file).read().splitlines()]
    start_row, start_col = find_point(grid, "S")
    end_row, end_col = find_point(grid, "E")
    walls = set(find_points(grid, "#"))

    savings = find_cheats(grid, start_row, start_col, end_row, end_col, walls)

    return sum(1 for saving in savings if saving >= 100)


def part_2(file: str = "Input/InputDay20P1.txt") -> int:
    grid = [list(line) for line in open(file).read().splitlines()]
    start_row, start_col = find_point(grid, "S")
    end_row, end_col = find_point(grid, "E")
    walls = set(find_points(grid, "#"))

    savings = find_cheats(
        grid, start_row, start_col, end_row, end_col, walls, max_cheat_time=20
    )

    return sum(1 for saving in savings if saving >= 100)


print(part_2())
```
